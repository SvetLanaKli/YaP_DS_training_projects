```python
def load_data(file_name):
    '''функция загрузки файлов'''
    try:
        folder_path = './'
        file_path = folder_path +'/'+ file_name + '.csv'
        data = pd.read_csv(file_path, parse_dates = ['date'])

    except:
        folder_path = 'https://link/datasets'
        file_path = folder_path +'/'+ file_name + '.csv'
        data = pd.read_csv(file_path, parse_dates = ['date'])
        
    return data
```

```python
def data_info(data_name):
    '''функция выводит информацию о данных'''
    print('\nИнформация о данных')
    display(data_name.info())

    print('\nДанные из выборки')
    display(data_name.head(5))
    display(data_name.tail(5))

    print(f'\nКоличество признаков: {data_name.shape[1]}')
    print(f'\nКоличество наблюдений: {data_name.shape[0]}')
    
    print('\nКоличество пропусков в данных')
    display(data_name.isna().sum().to_frame(name='quantity').query('quantity > 0').\
            style.background_gradient('coolwarm'))

    print(f'\nКоличество полных дубликатов в данных: {data_name.duplicated().sum()}')


    print('\nОписательная статистика ')
    display(data_name.describe())
```

```python
def recovery_mae(feed, concentrate, tail, recovery):
    '''функция рассчета эфективности обогащение и MAE'''
    df_recovery = (concentrate * (feed - tail)) / (feed * (concentrate - tail)) * 100
    mae = mean_absolute_error(df_recovery, recovery)
    print(f'MAE: {mae}')
```

```python
def metall_hist(list_metall, metall_name):
    '''функция вывода диаграммы концентрации металлов на различных этапах'''
    for item in list_metall:
        df_full[item].hist(alpha=0.9, figsize=(12,6), bins=30);
    plt.title(f'Концентрация металла {metall_name} на различных этапах');
    plt.legend(list_metall);
    plt.xlabel(f'интервалы значений признака')
    plt.ylabel('частота наблюдений')
    plt.show();
```

```python
def num_per_poor(df, name_metall, threshold, name_data='исходые данные'):
    '''функция подсчета количества наблюдений руды c процентным 
    содержанием менее заданного порога'''
    n = df[df[name_metall] < threshold].shape[0] 
    per = n / df.shape[0] 
    print(f'Количество наблюдений в {name_data} с содержанием золотой руды менее {threshold}%: {n}, что составляет {per:.2%}')
```

```python
def del_poor(df, name_metall, threshold):
    '''функция удаления наблюдений руды c процентным содержанием 
    менее заданного порога'''
    df = df[df[name_metall] > threshold]
    return df
```

```python
def smape(target, prediction):
    '''функция расчета sMAPE'''
    smape = abs(target - prediction) / ((abs(target) + abs(prediction)) * 0.5)
    smape = pd.DataFrame(smape).mean() * 100
    smape = smape.fillna(0)
    return smape
```

```python
def smape_total(rougher, final):
    '''функция расчета итоговой sMAPE'''
    smape_total = 0.25 * rougher + 0.75 * final
    print(f'Итоговое sMAPE: {smape_total:.4f}%')
    return smape_total
```

```python
def model(pipeline, features, target, name = 'rougher'):
    '''функция оценки качества модели методом кросс-валидации
    на входе: пайплайн, фичи и таргет
    на выходе: среднее значение метрики sMAPE'''
    model = cross_val_score(pipeline,
                            features, target,
                            scoring=make_scorer(smape, greater_is_better=False),
                            cv=CV,
                            n_jobs=-1
                           )
    print(f'sMAPE для {name}: {abs(model.mean()):.4}%')
    return abs(model.mean())
```

```python
def best_model(model, params, features, target, name, search='grid'):
    '''функция подбора гиперпараметров для заданной модели
    на входе: модель, параметры для подбора, фичи и таргет, название стадии,
                 способ подбора параметров, по умолчанию GridSearchCV
    на выходе: подобранная модель и значение искомой метрики sMAPE'''
    
    # сделаем пайплайн
    pipe = Pipeline([
    ('scale', StandardScaler()),
    ('model', model)
    ])
    
    if search == 'grid':
        searcher = GridSearchCV(pipe,
                                param_grid=params,
                                scoring=make_scorer(smape, greater_is_better=False),
                                cv=CV,
                                n_jobs=-1,
                                verbose=10)
    else:
        searcher = RandomizedSearchCV(pipe,
                                      param_distributions=params,
                                      scoring=make_scorer(smape, greater_is_better=False),
                                      cv=CV,
                                      n_jobs=-1,
                                      verbose= 10)
    
    # подбираем гиперпараметры
    search = searcher.fit(features, target)
    best_params.append({'model':search.best_estimator_,
                        #'model_params':search.best_params_, 
                        'name': name,   
                        'smape': abs(search.best_score_)})
    return best_params
```

### Функция для вычисления `sMAPE`.

Метрика SMAPE (symmetric mean absolute percentage error): вычисляется по следующей формуле:
    

 sMAPE $= \frac{1}{N}\sum_{i=1}^{N} \frac{|y_i - \hat{y_i}|}{{(|y_i| + |\hat{y_i}|)/2}} \times 100\% $

Итоговое sMAPE =25% $×$ sMAPE(rougher)+75% $×$sMAPE(final)
