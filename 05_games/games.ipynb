## Введение

### Описание проекта

**Исследование рынка компьютерных игр**

В распоряжении исторические данные до 2016 года из открытых источников о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 
Необходимо выявить закономерности, определяющие успешность игры. Отобрать данные актуальные для планирования рекламной компании на 2017 год. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

### Цель исследования

1. Определить успешность игры
2. Выявить, какие факторы влияют на успешность игры
3. Отобрать актуальные для планирования рекламной компании данные  
4. Составить портрет пользователя игр
5. Проверить гипотезы:
 * рейтинги разных платформ идентичны
 * рейтинги разных жанров различаются

### Ход исследования

Данные для исследования экспортирует из csv-файла.

О качестве данных информации нет, поэтому преже чем приступать к исследованию проведем обзор данных.

Также  необходимо будет проверить данные на наличие ошибок и их влияние на исследование. Далее на этапе предобработки рассмотрим возможность исправления ошибок и заполнения пропусков, которые не приведут к искажению результатов исследования.
После создадим дополнительные признаки, которые помогут в исследовании. И приступим непосредственно к исследованию.

Таким образом, в ходе проекта выполним следующие **этапы**:
* Обзор данных
* Предобработка данных
* Расчеты и добавление результатов в таблицу
* Проведение анализа

Займемся обработкой столбца `year_of_release`. Преобразуем его к типу целочисленному типу и проверим количество и возможности заполнения пропуcков.

```python
data['year_of_release'] = data['year_of_release'].astype('Int64') 
```

* Как видно, пропуски имеются по различным платформам и жанрам, какой то однозначной зависимости нет. Скорее всего, данные были не корректно выгружены из первисточников.

* В названии игр иногда содержится и обозначения года, но они заполнены не по одному формату, поэтому пытаться выбрать из названия игры год ее выпуска, считаю не целесообразным. 

* Заполнять год выпуска игры, используя информацию из датафрейма не совсем логично: допустим взять среднее, минимальное или первое, так как это может сильно исказить данные. 

Поэтому на текущем этапе оставим пропуски незаполненными, вернемся к этому вопросу если возникнет необходимость.

Займемся теперь самыми масштабными пропусками в нашем наборе данных:

* `critic_score` - оценка критиков (максимум 100)
* `user_score` - оценка пользователей (максимум 10)
* `rating` - рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
#сделаем срез по отсутсвующим рейтингам
data.query('critic_score.isna() & user_score.isna() & rating.isna()')
```

Как видно в данных более $6500$ строк с незаполненными рейтингами.

По данным Википедиии **ESRB** была основана в $1994$ году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»). К началу $2003$ года рейтинги ESRB получили более $8 000$ продуктов от 350 издателей.

Проверим были ли рейтинги до $1994$ года

```python
data[data['year_of_release'] < 1994]['rating'].count()
```

Действительно до $1994$ года возрастного рейтинга у игр не было, и как видно из данных только $4$ игры их получили после введения рейтингования.
При этом игр до $1994$ в данных представлено всего $365$ шт. 

Причиной пропуска данных скорее всего служит несколько факторов:
 * `ESRB` основана в $1994$ году, и соответственно игры до этого момента не рейтинговались
 * `ESRB` рейтингует игры рынка Северной Америки и Канады, в Европе и Японии имеются свои рейтинги

Скорее всего, пропуски возникли по причине их отсутсвия в источнике данных. 

С пропусками поступим следующим образом:
* `critic_score` заполнение рейтинга на текущем этапе не целесообразно. Приведем признак к целочисленному типу;
* `user_score` кроме `NaN` содержит  `tbd` (подлежит определению), эти данные были умышленно заполнены таким значением, заменим его на `NaN`, чтобы было единообразие. После преобразуем столбец к типу `float`;
* `raiting` - переменная содержащая категории, заполним пропуски значением `unknown`.

```python
#заменим tbd на NaN
data.loc[data['user_score'] == 'tbd', 'user_score'] = 'NaN'
data['user_score'] = data['user_score'].astype('float')
#проверим, что получилось
data['user_score'].unique()
```

```python
#заменим на unknown
data.loc[data['rating'].isna(), 'rating'] = 'unknown'
#проверим, что получилось
data['rating'].unique()
```

Заменим устаревшие рейтинги
* `K-A` ("Kids to Adults") на `E` ("Everyone")
* `AO` ("Adults Only") на `M` ("Mature")

Замену `EC` на `E` делать не будем, так как рейтинг изменен только в 2018 году.

```python
#заменим на unknown
data.loc[data['rating'] == 'K-A', 'rating'] = 'E'
data.loc[data['rating'] == 'AO', 'rating'] = 'M'
#проверим, что получилось
data['rating'].unique()
```

```python
#преобразуем critic_score в int
data['critic_score'] = data['critic_score'].astype('Int64')
```

8. Рассмотрим общее распределение игр по жанрам. Выясним, какие жанры являются самыми прибыльными и распространенными.


```python
#построим график распределения суммарных и медианных продаж, количества игр в разрезе жанров

fig = plt.figure()

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)


(data_shot
 .pivot_table(index='genre', values='sales', aggfunc=['sum', 'count'])
 .droplevel(1, axis=1)
 .sort_values(by='sum', ascending=False)
 .plot(kind='bar', 
       xlabel='жанры', ylabel='продажи',
       ax=ax_1)
)

(data_shot
 .pivot_table(index='genre', values='sales', aggfunc=['median', 'mean'])
 .droplevel(1, axis=1)
 .sort_values(by='median', ascending=False)
 .plot(kind='bar', 
       xlabel='жанры', ylabel='продажи',
       ax=ax_2)
)

ax_1.set_title('Распределение суммарных продаж и количества игр по жанрам')
ax_2.set_title('Распределение медианных продаж по жанрам')


fig.set_figwidth(20)
fig.set_figheight(6)

plt.show()
```

## Проверка гипотез

Необходимо проверить следующие гипотезы:

1. Средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые;
2. Средние пользовательские рейтинги жанров `Action` (англ. «действие», экшен-игры) и `Sports` (англ. «спортивные соревнования») разные.

1. Проверка гипотез всегда начинается с формулировки нулевой $H_0$ и альтернативной $H_1$:

  * $H_0$: Средние  пользовательские рейтинги (`user_score`) платформы `XOne` $=$ средним пользовательстким рейтингам (`user_score`) платформы `PC`
  * $H_1$: Средние  пользовательские рейтинги (`user_score`) платформы `XOne` $\neq$  средним пользовательстким рейтингам (`user_score`) платформы `PC`


Зададим критический уровень значимости alpha = 0.01.

```python
#создадим выборки для тестирования гипотез, удалим пропуски в данных
user_score_xone = data_shot[data_shot['platform'] == 'XOne']['user_score'].dropna()
user_score_pc = data_shot[data_shot['platform'] == 'PC']['user_score'].dropna()
```

```python
#проверим как распределены выборки
plt.figure(figsize=(6,4))
user_score_xone.plot(kind='hist', alpha=1, bins=10);
user_score_pc.plot(kind='hist', alpha=0.5, bins=10);
plt.title('Распределение выборок "user_score"');
plt.show();
```

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой.
Для проверки на равенство используем более строгий критерий значимости 0.01

```python
# results = вызов метода ttest_ind для проверки гипотезы равенства средних двух ген.совокупностей
results = st.ttest_ind(user_score_xone, user_score_pc, equal_var=False)

# alpha = значение уровня значимости
alpha = 0.01

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**: По всей видимости, средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые.

 2. Проверка гипотез всегда начинается с формулировки нулевой $H_0$ и альтернативной $H_1$:

  * $H_0$: Средние  пользовательские рейтинги (`user_score`) жанра `Action` $=$ средним пользовательстким рейтингам (`user_score`) жанра `Sports`
  * $H_1$: Средние  пользовательские рейтинги (`user_score`) жанров `Action` $\neq$  средним пользовательстким рейтингам (`user_score`) жанра `Sports`


Зададим критический уровень значимости alpha = 0.01

```python
#создадим выборки для тестирования гипотез, удалим пропуски в данных
user_score_action = data_shot[data_shot['genre'] == 'Action']['user_score'].dropna()
user_score_sports = data_shot[data_shot['genre'] == 'Sports']['user_score'].dropna()
```

```python
#проверим как распределены выборки
plt.figure(figsize=(6,4))
user_score_action.plot(kind='hist', alpha=1, bins=10);
user_score_sports.plot(kind='hist', alpha=0.5, bins=10);
plt.title('Распределение выборок "user_score"');
plt.show();
```

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой.
Для проверки на равенство используем более строгий критерий значимости 0.01

```python
# results = вызов метода ttest_ind для проверки гипотезы равенства средних двух ген.совокупностей
results = st.ttest_ind(user_score_action, user_score_sports, equal_var=False)

# alpha = значение уровня значимости
alpha = 0.01

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** Скорее всего, средние пользовательские рейтинги жанров `Action` и `Sports` неодинаковые.
