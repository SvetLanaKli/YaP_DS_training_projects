## Введение

### Цель исследования

Построение модели для опредсказания темпетуры стали по данным , предоставленным заказчиком с целью минимизации энергозатрат, сочетающей в себе оптимальное соотношение качества, скорости предсказания и времени обучения модели.

### План исследования

1. Обзор и исследование данных
* анализ на пропуски, дубликаты, аномалии
* выявление зависимых признаков
* удаление явных аномалий
2. Подготовка данных
* приведение наименований признаков к хорошему стилю
* заполнение пропусков
* генерация дополнительных признаков
* удаление неинформативных данных
* объединение данных
* разделение выборок на обучающию и тестовую
3. Построение моделей машинного обучение на основе алгоритмов
* `Linear/Ridge`
* `RandomForest`
* `LightGBM`
* `CatBoost`

    Обучение моделей с поиском оптимальный параметров будем проводить с помощью `GridSearchCV/RandomizedSearchCV`, использующие кросс-валидацию. Для оптимизации кода, и устранения проблемы утечки данных будем использоваться `Pipeline`
    
    Выбор наиболее соотвествующей параметрам заказчика модели
4. Тестирование выбранной модели
5. Оформление выводов

Данные состоят из файлов, полученных из разных источников и содержат информацию: 
* об электродах;
* о подаче сыпучих материалов (объём  и время);
* о продувке сплава газом;
* о результатах измерения температуры;
* о проволочных материалах (объём и вреямя).

Во всех файлах содержится информация о номере партии. В файлах может быть несколько строк с одинаковым значением номера партии, они соответствуют разным итерациям обработки.

```python
def data_load(file_name):
    '''функция загрузки файлов'''
    try:
        folder_path = '/datasets'
        file_path = folder_path +'/'+ file_name + '.csv'
        data = pd.read_csv(file_path, parse_dates=True)
    except:
        folder_path = 'link/datasets'
        file_path = folder_path +'/'+ file_name + '.csv'
        data = pd.read_csv(file_path, parse_dates=True)
        
    return data

def data_info(data_name):
    '''функция выводит информацию о данных'''
    
    print('\n\nДанные из выборки\n')
    display(data_name.head(5))
    display(data_name.tail(5))

    print(f'\n\nКоличество признаков: {data_name.shape[1]}, Количество наблюдений: {data_name.shape[0]}\n')
    
    print('\nИнформация о данных\n')
    info_list = []
    for item in data_name.columns:
        info_list.append({
            'Признак': item,
            'Тип данных': data_name.dtypes[item],
            'Уникальных значений': data_name[item].nunique(),
            '%Пропусков': data_name[item].isna().mean()*100
        })
    display(pd.DataFrame(info_list).style.bar(subset=['%Пропусков']))
   

    print(f'\n\nКоличество полных дубликатов в данных: {data_name.duplicated().sum()}')

    print('\n\nОписательная статистика ')
    display(data_name.describe().T)
```

```python
def data_corr(data_name, figsize=(6,6)):
    '''функция выводит матрицу корреляции признаков'''
    sns.set(style="white", palette="Spectral")
    mask = np.triu(np.ones_like(data_name.corr(), dtype=bool));
    plt.figure(figsize=figsize)
    
    sns.heatmap(data_name.corr(), mask=mask,
                annot=True,
                fmt=".2f",
                linewidth=.5).set(title='Матрица корреляции признаков');
    plt.show();
```

```python
def box_hist_plot(data, columns, bins=100):
    '''визуализация распределения'''
    sns.set(style="whitegrid", palette="Spectral")

    # построим полученное распределение
    f, (ax_box, ax_hist) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize = (6,8))
    f.suptitle(f'Распределение признака {columns}')
    # ящик с усами
    sns.boxplot(data=data, x=columns, ax=ax_box)
    ax_box.set_title('Диаграмма размаха')
    
    sns.histplot(data=data, x=columns, ax=ax_hist, bins=bins, kde=True)
    ax_hist.set_title('Распределение')
    # построим границы доверительного интервала
    ymin, ymax = ax_hist.get_ylim()
    x_lower = data[columns].quantile(0.25)
    x_upper = data[columns].quantile(0.75)
    ax_hist.vlines(x=[x_lower, x_upper], ymin=ymin, ymax=ymax, ls='--')

    plt.show()
```

```python
def best_model(model, params, features, target, name, scaler=False, grid=True):
    '''функция подбора гиперпараметров для заданной модели
    на входе: модель, параметры для подбора, фичи и таргет,
              категориальные признаки, числовые признаки,
              способ подбора параметров, по умолчанию RandomozedSearchCV
    на выходе: подобранная лучшая модель и значение искомой метрики MAE'''
      
    if scaler:
        print('scaler')
        pipe = Pipeline(steps=[
            ('impute', SimpleImputer(fill_value=0)),
            ('scaler', StandardScaler()),
            ('model', model)])
    else:
        pipe = Pipeline(steps=[
            ('impute', SimpleImputer(fill_value=0)),
            ('model', model)])
         
    if grid:
        print('Grid')
        searcher = GridSearchCV(pipe,
                                param_grid=params,
                                scoring='neg_mean_absolute_error',
                                cv=CV,
                                n_jobs=-1,
                                verbose=10)    
    else:
        # подбор гиперпараметров
        searcher = RandomizedSearchCV(pipe,
                                      param_distributions=params,
                                      scoring='neg_mean_absolute_error',
                                      random_state=RS,
                                      n_iter=200,
                                      n_jobs=-1,
                                      verbose= 10,
                                      cv = CV,
                                      return_train_score=True)

    search = searcher.fit(features, target)
    
    # вычисляем время работы с лучшими параметрами
   
    time_start = dt.now()
    search = searcher.fit(features, target)
    time_search = (dt.now() - time_start)
    
    # вычисляем время работы с лучшими параметрами
    pipe.set_params(**search.best_params_)
    
    time_start = dt.now()
    pipe.fit(features, target)
    time_fit = (dt.now() - time_start)
    
    time_start = dt.now()
    pipe.predict(features)
    time_predict = (dt.now() - time_start)
    
    list_best_params.append({'model_name': name,
                             'model': search.best_estimator_[-1],
                             'model_params': search.best_params_,
                             'MAE': abs(search.best_score_),
                             'time_search': time_search,
                             'time_fit': time_fit,
                             'time_predict': time_predict})
```

```python
df_arc['heating_sec'] = ((df_arc['heating_end'] - df_arc['heating_begin'])
                                         .apply(lambda x: x.total_seconds())
                                         .astype('int'))
```

Вспоминая термофизику и физику в целом, сгенерируем новый признак:

* полную энергию $ S = \sqrt{P^2 + Q^2}$

```python
df_arc['work'] = df_arc['heating_sec'] * df_arc['power_active']
df_arc['power_full'] = (df_arc['power_active'] ** 2 + df_arc['power_reactive'] ** 2) ** 0.5
```
