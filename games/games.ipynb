С пропусками поступим следующим образом:
* `critic_score` заполнение рейтинга на текущем этапе не целесообразно. Приведем признак к целочисленному типу;
* `user_score` кроме `NaN` содержит  `tbd` (подлежит определению), эти данные были умышленно заполнены таким значением, заменим его на `NaN`, чтобы было единообразие. После преобразуем столбец к типу `float`;
* `raiting` - переменная содержащая категории, заполним пропуски значением `unknown`.

```python
#заменим на unknown
data.loc[data['rating'].isna(), 'rating'] = 'unknown'
#проверим, что получилось
data['rating'].unique()
```

```python
#отберем лидеров продаж среди платформ
top_sales = data\
.pivot_table(index='platform', values='sales', aggfunc='sum')\
.sort_values(by='sales', ascending=False)\
.reset_index()\
.head(10)
```

```python
#построим графики продаж по годам в разрезе платформ ТОП10
data[data['platform'].isin(top_sales.platform)]\
.pivot_table(index='year_of_release', columns='platform', values='sales', aggfunc='sum')\
.plot(kind='line', title='Продажи в разрезе ТОП10 платформ по годам',
     figsize=(12,8), xlabel='годы', ylabel='продажи, млн.');
```

```python
#рассмотрим период жизни платформ
for i in range(len(top_sales.platform)):
    data[data['platform'] == top_sales.platform[i]].groupby('year_of_release')\
                .agg('sum').plot(kind='bar',  y='sales', title=top_sales.platform[i])
```

```python
#функция вывода гистограмм по регионам в разрезе параметра plot_index

def region_plot (plot_index):
    '''функция построения гистограмм распределения продаж в разрезе ТОП-5 
       заданного параметра по каждому региону исследования'''
    
    fig, axs = plt.subplots(1, 3, figsize=(18,6))
    
    for i in range(len(regions)):
        
        df_i = data_shot.groupby(plot_index)[regions[i]].sum().reset_index().sort_values(by=regions[i], ascending=False).head(5)
        #print(df_i)
        axs[i].bar(df_i[plot_index], df_i[regions[i]], color='maroon')
        axs[i].set_title(regions[i])
    
    plt.show()      
```

```python
#функция вывода круговой диаграммы по регионам в разрезе параметра plot_index

def region_plot_pie (plot_index):
    '''функция построения гистограмм распределения продаж в разрезе ТОП-5 
       заданного параметра по каждому региону исследования'''
    
    fig, axs = plt.subplots(1, 3, figsize=(18,6))
    
    for i in range(len(regions)):
        
        df_i = (data_shot
                .groupby(plot_index)[regions[i]]
                .agg('sum')
                .reset_index()
                .sort_values(by=regions[i], ascending=False)
                .head(5)
                )
        sale_other = (data_shot[regions[i]].sum() - df_i[regions[i]].sum()).round(2)
        df_i.loc[len(df_i.index)] = ['Others', sale_other]
        #print(df_i)
        axs[i].pie(df_i[regions[i]], labels=df_i[plot_index], autopct='%1.0f%%', pctdistance=0.7)
        axs[i].set_title(regions[i])
    
    plt.show()
```

## 5. Проверка гипотез

**Задачи:**

Проверить следующие гипотезы:

1. Средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые;
2. Средние пользовательские рейтинги жанров `Action` (англ. «действие», экшен-игры) и `Sports` (англ. «спортивные соревнования») разные.

1. Проверка гипотез всегда начинается с формулировки нулевой $H_0$ и альтернативной $H_1$:

  * $H_0$: Средние  пользовательские рейтинги (`user_score`) платформы `XOne` $=$ средним пользовательстким рейтингам (`user_score`) платформы `PC`
  * $H_1$: Средние  пользовательские рейтинги (`user_score`) платформы `XOne` $\neq$  средним пользовательстким рейтингам (`user_score`) платформы `PC`


Зададим критический уровень значимости alpha = 0.01.

```python
#создадим выборки для тестирования гипотез, удалим пропуски в данных
user_score_xone = data_shot[data_shot['platform'] == 'XOne']['user_score'].dropna()
user_score_pc = data_shot[data_shot['platform'] == 'PC']['user_score'].dropna()
```

```python
#проверим как распределены выборки
user_score_xone.plot(kind='hist', alpha=1, bins=10);
user_score_pc.plot(kind='hist', alpha=0.5, bins=10);
```

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой.
Для проверки на равенство используем более строгий критерий значимости 0.01

```python
# results = вызов метода ttest_ind для проверки гипотезы равенства средних двух ген.совокупностей
results = st.ttest_ind(user_score_xone, user_score_pc, equal_var=False)

# alpha = значение уровня значимости
alpha = 0.01

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**: По всей видимости, средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые.

 2. Проверка гипотез всегда начинается с формулировки нулевой $H_0$ и альтернативной $H_1$:

  * $H_0$: Средние  пользовательские рейтинги (`user_score`) жанра `Action` $=$ средним пользовательстким рейтингам (`user_score`) жанра `Sports`
  * $H_1$: Средние  пользовательские рейтинги (`user_score`) жанров `Action` $\neq$  средним пользовательстким рейтингам (`user_score`) жанра `Sports`


Зададим критический уровень значимости alpha = 0.01

```python
#создадим выборки для тестирования гипотез, удалим пропуски в данных
user_score_action = data_shot[data_shot['genre'] == 'Action']['user_score'].dropna()
user_score_sports = data_shot[data_shot['genre'] == 'Sports']['user_score'].dropna()
```

```python
#проверим как распределены выборки
user_score_action.plot(kind='hist', alpha=1, bins=10);
user_score_sports.plot(kind='hist', alpha=0.5, bins=10);
```

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой.
Для проверки на равенство используем более строгий критерий значимости 0.01

```python
# results = вызов метода ttest_ind для проверки гипотезы равенства средних двух ген.совокупностей
results = st.ttest_ind(user_score_action, user_score_sports, equal_var=False)

# alpha = значение уровня значимости
alpha = 0.01

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** Скорее всего, средние пользовательские рейтинги жанров `Action` и `Sports` неодинаковые.

На планируемый $2017$ год следует обратить внимание на игры для следующих платформ:

 *	`PS4`
 *  `XOne`
 *	`3DS`
 *  `WiiU`
 
 Наиболее перспективными жанрами игр являются:
* `Shooter`
* `Sports` 
* `Platform`
* `Role-Playing`
* `Action`

Для пользователей Северной Америки и Европы наиболее перспективными играми будут являться игры для `PS4` в жанрах `Shooter`
и `Sports`.  

Для жителей Японии, не смотря на постепенное снижение спроса, актуальными остаются игры для `3DS`  в жанре `Role-Playing`.
При выборе игр пользоватетели больше ориентируются на оценки критиков, чем на пользовательские рейтинги.
